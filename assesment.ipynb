```python
import pandas as pd
```

```python
#Reading the csv file 1 and storing it 
df1 = pd.read_csv('./data/factbook.csv',sep=';')
#df1.describe()

# Dataframe1 preprocessing
df1 = df1.iloc[1:]
```

```python
#Reading csv file 2 storing it
df2 = pd.read_csv('./data/factbook-2.csv',sep=';')
#df2.describe()

# Dataframe1 preprocessing
df2 = df2.iloc[1:]
```

```python
#Merging two file using ConcatFunction
merged_df =pd.concat([df1,df2])
merged_df
#eliminating duplicates
merged_df.drop_duplicates(ignore_index=True)
```

```python
# Writing it into file in temp/temp.csv
merged_df.to_csv('./temp/temp.csv')
```

```python
# getting data from temp and storing it
temp_data = pd.read_csv('./temp/temp.csv')
```

```python
#Removing Null Values
temp_data.fillna(0.00,inplace=True)
```

```python
# Creating group by key
temp_data['group_by_key'] = temp_data['Country'].str[0]
temp_data
```

```python
#Removing Country and Unnamed
temp_data = temp_data.drop(columns=['Country'])
temp_data = temp_data.loc[:, ~temp_data.columns.str.contains('^Unnamed')]
```

```python
temp_data
```

```python
#Processing Averge to all the fields for each group 
grouped_df = temp_data.groupby('group_by_key').mean()
grouped_df
```

```python
#inserting the dataframe to the output file
grouped_df.to_csv('./output/output.csv')
```
